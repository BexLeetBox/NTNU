```python
import numpy as np
import matplotlib.pyplot as plt

start = -1
slutt = 1
dx = 0.00001

def f(x):
    return x**4*(np.sin(x)*np.cos(x)+1)

range = np.arange(start,slutt,0.0001)

plt.plot(range,f(range))

plt.grid()


plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Figure of the original function f(x)")

x = start
Areal = 0

while x<slutt:
    Areal += f(x)*dx
    x+=dx

print(Areal)
```
